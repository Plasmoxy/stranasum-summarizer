```python

import pandas as pd 
raw = pd.read_csv('testset_evaluation_data.csv')
```

```python
def strip_special_tokens(text):
    text = text.replace("<SOS>", "").replace("<EOS>", "")
    text = text.replace("<sos>", "").replace("<eos>", "")
    text = text.replace("<unk>", "?").replace("<UNK>", "?")
    text = text.strip()
    return text

df = raw.copy()
df['article'] = df['article'].apply(strip_special_tokens)
df['summary'] = df['summary'].apply(strip_special_tokens)
df['predicted'] = df['predicted'].apply(strip_special_tokens)
df
```
